```python
import pandas as pd
import os

desktop = os.path.join(os.path.expanduser("~"), "Desktop")
file_path = os.path.join(desktop, "douyin.csv")

df = pd.read_csv(file_path)
print(" File read successfully！")
df.head()
```

```python
df.head()

print("Column names:")
print(df.columns.tolist())

print("\nBasic data information:")
df.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

df['Gender'].value_counts().plot(kind='pie', autopct='%.1f%%', figsize=(6,6), colors=['#ff9999','#66b3ff'])
plt.title('User Gender Distribution')
plt.show()

df['Age'].value_counts().sort_index().plot(kind='bar', figsize=(10,5), color='#ffcc99')
plt.title('User Age Distribution')
plt.ylabel('User Count')
plt.show()

df['Purchase_Frequency'].plot(kind='hist', bins=15, figsize=(10,5), color='#99ff99', edgecolor='black')
plt.title('Spending Amount Distribution')
plt.xlabel('Spending Amount')
plt.ylabel('User Count')
plt.show()
```

```python
age_count = df['Age'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
age_count.plot(kind='bar', color='orange')
plt.title('Age Distribution of Douyin E-commerce Users')
plt.xlabel('Age')
plt.ylabel('User Count')
plt.xticks(rotation=60)  
plt.show()
```
